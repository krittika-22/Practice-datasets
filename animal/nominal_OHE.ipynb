```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## CREATE A SIMPLE DATAFRAME:

df=pd.DataFrame({
    'color':['Red','Blue','Green','Green','Red','Blue']
})

df.head()
```

```python
## CREATE AN INSTANCE OF ONE HOT ENCODER:

encoder=OneHotEncoder()


## PERFORM FIT AND THEN TRANSFORM:

encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encoder_df
```

```python
## for new data

encoder.transform([['Blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

# ANIMAL DATASET

```python
import pandas as pd
```

```python
df=pd.read_csv('Animal Dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
## APPLY OHE WITH SCIKIT LEARN:

from sklearn.preprocessing import OneHotEncoder
```

```python
categorial_columns=['Animal','Color','Diet','Habitat']
categorial_columns

```

```python
encoder=OneHotEncoder()

encoded=encoder.fit_transform(df[categorial_columns]).toarray()

encoded_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encoded_df


```

```python
encoded_df.head()
```

```python
df_final=pd.concat([df.drop(columns=categorial_columns),encoded_df],axis=1)
df_final
```
