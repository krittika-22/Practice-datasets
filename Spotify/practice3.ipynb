```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# ===============================
# Step 1: Authenticate with Spotify
# ===============================
client_id = "YOUR_CLIENT_ID"
client_secret = "YOUR_CLIENT_SECRET"

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id=client_id,
    client_secret=client_secret
))

# ===============================
# Step 2: Load your dataset
# ===============================
df = pd.read_csv("dataset.csv")   # change filename if needed

# ===============================
# Step 3: Work only with first 50 tracks
# ===============================
subset = df.head(50).copy()

release_years = []

for track_id in subset['track_id']:
    try:
        track_info = sp.track(track_id)
        release_date = track_info['album']['release_date']
        year = release_date.split("-")[0]
        release_years.append(year)
    except:
        release_years.append(None)

# ===============================
# Step 4: Add release years to subset
# ===============================
subset['release_year'] = release_years

# ===============================
# Step 5: Show results
# ===============================
print(subset[['track_id', 'release_year']].head(10))

# ===============================
# Step 6: Count how many tracks per year
# ===============================
year_counts = subset['release_year'].value_counts().sort_index()
print(year_counts)
```

```python
tracks_per_year = df['release_year'].value_counts().sort_index()
print(tracks_per_year)
```

```python
####Which track has the highest popularity?
df.sort_values(by='popularity',ascending=False).head(1)[['track_name','artists','popularity']]
```

```python
##Genre with highest average popularity

avg_popularity=df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False)
print(avg_popularity)
```

```python
avg_popularity=df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False)
print(avg_popularity)

avg_popularity.head(10).plot(kind='bar',figsize=(10,5),color='green')
plt.title("TOP 10 GENRES WITH AVERAGE POPULARITY")
```

```python
## Explicit vs Non explicit popularity:

popularity=(df.groupby('explicit')['popularity'].mean())
print(popularity)
popularity.plot(kind='bar',title="EXPLICIT VS NON EXPLICIT POPULARITY")
plt.xlabel("Explicit content")
plt.ylabel("Average popularity")
```

```python
##Happiness (Valence) in Popular vs Less Popular

print(df.groupby(df['popularity'] > 50)['valence'].mean())

df.groupby(df['popularity'] > 50)['valence'].mean().plot(kind='bar', title="Valence: Popular vs Less Popular",color='pink')
```
